```python
import pandas as pd
pathSuperstore = r'Sample - Superstore.xlsx'
superstoreData = pd.read_excel(pathSuperstore)
```

```python
superstoreData.columns
```

```python
# Create the count, mean, standard deviation (std), minimun (min), and maximum (max) of the sales and profits for region in each year (20 pts)
superstoreData['Year'] = superstoreData['Order Date'].dt.year  
superstoreDataGBY = superstoreData.groupby(['Region', 'Year'])
superstoreDataGBY[['Sales','Profit']].agg(['count','mean','std','min','max'])
```

```python
#Create a pie chart to illustrate the total sales per region (20 pts)
##The pie chart will have 4 slices representing the 4 regions
##The total sales will correspond to the size of the slices
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#superstoreData['Year'] = superstoreData['Order Date'].dt.year  
superstoreDataGBR = superstoreData.groupby(['Region'])
y = superstoreDataGBR.sum(numeric_only = True).Sales
labels = np.array(superstoreDataGBR.mean(numeric_only = True).index)

plt.pie(y, labels=labels, autopct='%1.2f%%', shadow=True, explode=[0.1,0,0,0])
plt.show() # Display the pie chart
```

```python
def top(df, column, n=5):
    return df.sort_values(by=column)[-n:]
```

```python
#Return the top 3 orders in each region based on the total sales (20 pts)
##You can use a user-defined function following the lecture notes

#superstoreData['Year'] = superstoreData['Order Date'].dt.year  
superstoreDataGBR = superstoreData.groupby(['Region'])
superstoreDataGBR.apply(top, column='Sales', n=3) 
```
